# Cropping NDVI Raster to Cologne Region
Assuming we have a Cologne boundary vector layer (e.g., named cologne_boundary):

```python
import processing

# Names of your layers
raster_layer_name = "ndvi_cologne_reprojected copy"
mask_layer_name = "cologne_boundary"  # Change to your actual boundary layer name

# Get layers
project = QgsProject.instance()
raster_layer = [lyr for lyr in project.mapLayers().values() if lyr.name() == raster_layer_name][0]
mask_layer = [lyr for lyr in project.mapLayers().values() if lyr.name() == mask_layer_name][0]

# Set output path
output_path = os.path.abspath(os.path.join("..", "data", "outputs", "ndvi_cologne_only.tif"))

# Run the clip by mask algorithm
params = {
    'INPUT': raster_layer,
    'MASK': mask_layer,
    'OUTPUT': output_path,
    'CROP_TO_CUTLINE': True,
    'KEEP_RESOLUTION': True
}
processing.run("gdal:cliprasterbymasklayer", params)

print(f"NDVI raster cropped to Cologne region saved at: {output_path}")


processing.run("gdal:cliprasterbymasklayer", {'INPUT':'/Users/rk/Sync/courses/correleraid/cologne_green_project/data/outputs/ndvi_cologne_reprojected.tif','MASK':'/Users/rk/Sync/courses/correleraid/cologne_green_project/data/boundaries/Stadtbezirk_18/Stadtbezirk.shp','SOURCE_CRS':None,'TARGET_CRS':None,'TARGET_EXTENT':None,'NODATA':None,'ALPHA_BAND':True,'CROP_TO_CUTLINE':True,'KEEP_RESOLUTION':False,'SET_RESOLUTION':False,'X_RESOLUTION':None,'Y_RESOLUTION':None,'MULTITHREADING':False,'OPTIONS':None,'DATA_TYPE':0,'EXTRA':'','OUTPUT':'TEMPORARY_OUTPUT'})
```
