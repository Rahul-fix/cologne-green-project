# Step 1: Import Required Libraries
Import PyQGIS and processing modules for raster reprojection.

```python
import os
from qgis.PyQt.QtWidgets import QApplication
import sys
from qgis.core import QgsProject, QgsRasterLayer, QgsCoordinateReferenceSystem
import processing

# Create a QApplication instance if one does not exist
app = QApplication.instance()
if app is None:
    app = QApplication(sys.argv)

# Path to your QGIS project file
project_path = "cologne_green_analysis.qgz"  # or "cologne_analysis.gpkg" if using GeoPackage

# Load the project
QgsProject.instance().read(project_path)

# Get all layers in the project
layers = QgsProject.instance().mapLayers().values()

# List all layers and their IDs
for layer in layers:
    print(f"Name: {layer.name()}, ID: {layer.id()}")

# l1 = layer.id()
# print(f"Using layer ID: {l1}")
```

# Step 2: Set Output Directory
Create and set the output directory for saving reprojected files.

```python
# output_dir = 'data/outputs/'
output_dir = '../data/out/'
os.makedirs(output_dir, exist_ok=True)
```

# Step 3: Define Target CRS
Set the target CRS for Cologne, Germany (EPSG:25832).

```python
target_crs = QgsCoordinateReferenceSystem('EPSG:25832')
```

# Step 4: Reproject Raster Layer
Reproject a raster layer and save the output to the output directory.

```python
# Replace 'your_layer_id' and 'your_output_filename.tif' as needed
layer_id = 'T31UGS_20250403T104041_B04_10m_45ef9821_2878_44dc_91ba_00b5b376ecb9'
output_path = os.path.join(output_dir, 'B04_reprojected.tif')
layer = QgsProject.instance().mapLayer(layer_id)
params = {
    'INPUT': layer,
    'TARGET_CRS': target_crs,
    'RESAMPLING': 0,
    'OUTPUT': output_path
}
result = processing.run('gdal:warpreproject', params)
reprojected_layer = QgsRasterLayer(output_path, 'B04_reprojected')
QgsProject.instance().addMapLayer(reprojected_layer)
```

# Step 5: Reproject Another Raster Layer
Repeat for another raster layer.

```python
# Replace 'your_layer_id' and 'your_output_filename.tif' as needed
layer_id = 'T31UGS_20250403T104041_B08_10m_7d7b6d9b_2b7e_43a5_b3cf_76fe0f7cdaa4'
output_path = os.path.join(output_dir, 'B08_reprojected.tif')
layer = QgsProject.instance().mapLayer(layer_id)
params = {
    'INPUT': layer,
    'TARGET_CRS': target_crs,
    'RESAMPLING': 0,
    'OUTPUT': output_path
}
result = processing.run('gdal:warpreproject', params)
reprojected_layer = QgsRasterLayer(output_path, 'B08_reprojected')
QgsProject.instance().addMapLayer(reprojected_layer)
```

# Step 6: Reproject Vector Layer (if needed)
Use this if your vector layer is not already in EPSG:25832.

```python
# Replace 'your_vector_layer_id' and 'your_output_filename.gpkg' as needed
layer_id = 'stadtviertel_2a8f4ac9_840b_47aa_86ac_da0cb9bdc8fc'
output_path = os.path.join(output_dir, 'stadtviertel_reprojected.gpkg')
layer = QgsProject.instance().mapLayer(layer_id)
params = {
    'INPUT': layer,
    'TARGET_CRS': target_crs,
    'OUTPUT': output_path
}
result = processing.run('qgis:reprojectlayer', params)
reprojected_layer = QgsVectorLayer(output_path, 'stadtviertel_reprojected', 'ogr')
QgsProject.instance().addMapLayer(reprojected_layer)
```
